```python
import matplotlib.pyplot as plt
import numpy
```

# Zadanie 12.1. 
W pliku `rysowanie1.csv` są zapisane wartości 2 funkcji. Narysuj obie te funkcje
na jednym wykresie. Na podstawie wykresu ustal, która z nich jest monotoniczna.

```python
input_file = numpy.loadtxt('pliczki/rysowanie1.csv', delimiter=",", skiprows=1)
x, f1, f2 = numpy.transpose(input_file)
```

```python
plt.plot(x, f1, label="f1")
plt.plot(x, f2, label="f2")
plt.legend()
plt.show()
```

# Zadanie 12.2. 
W pliku `rysowanie2.csv` są zapisane zyski (fikcyjnych) firm w kolejnych latach.
Porównaj je ze sobą. Określ, która firma radziła sobie w podanym okresie najlepiej, a która najgorzej, podaj jakieś prognozy na przyszłość.

```python
input_file = numpy.loadtxt('pliczki/rysowanie2.csv', delimiter=",", skiprows=2)
rok, heavy_coal_inc, shady_power, dysonspherics_gmbh = numpy.transpose(input_file)
```

```python
plt.plot(rok, heavy_coal_inc, label="heavy coal inc", marker=".")
plt.plot(rok, shady_power, label="shady power", marker=".")
plt.plot(rok, dysonspherics_gmbh, label="dysonspherics gmbh", marker=".")
plt.xlabel("Rok")
plt.ylabel("Zysk (mln euro)")
plt.title("Zyski firm heavy coal inc, shady power, dysonspherics gmbh w kolejnych latach")
plt.grid()
plt.legend()
plt.show()
```

# Zadanie 12.3
W pliku `rysowanie3.csv` zapisany jest wiek kolonistów, którzy być może będą mieszkali w pewnej kolonii na Wenus w roku 2248. Określ, ile osób znajduje się w każdym z przedziałów wiekowych: $[0, 10), [10, 20), [20, 30), . . . , [80, 90), [90, 100].$

```python
input_file = numpy.loadtxt('pliczki/rysowanie3.csv', delimiter=",", skiprows=1)
wiek = numpy.transpose(input_file)[1]
wiek
```

```python
plt.hist(wiek, bins=10)
plt.ylabel("Liczba kolonistów")
plt.xlabel("Wiek")
plt.xticks([i for i in range(0, 101, 10)])
plt.grid()
plt.title("")
plt.show()
```

W pliku `rysowanie5.csv` zapisana jest prognozowana na podstawie symulacji częstość wystąpienia zakażenia szpitalnego (w procentach) w 164 szpitalach w pewnym regionie w kolejnych dniach od przyjęcia pierwszej zakażonej osoby w jednym ze szpitali (na podstawie `https://journals.plos.org/ploscompbiol/article?id=10.1371/journal.pcbi.1008442`).

Czy częstości wystąpienia są zbliżone w każdym ze szpitali?
