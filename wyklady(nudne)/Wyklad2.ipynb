## PrzykÅ‚ad
$$f(x)=\left\{ \begin{array}{rl}
2x & \textrm{dla $x > 0$} \\ 3 & \textrm{dla $x \leq 0$}
\end{array} \right.$$

#### Prosto (skalarnie)

```python
def f_skalar(i):
    if i > 0:
        return 2*x
    return 3
def fL(v):
    return [f_skalar(j) for j in v]
fL(x)
```

#### Wektorowo

```python
def f_wektor(v):
    return 3.0*(v <= 0) + 2 * v * (v > 0)
f_wektor(x)
```
